```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

#### Підготуємо наші датасети 

```python
data_folder = "data/"

data_2d = pd.read_csv(data_folder + "data_2d.csv")
data_2d = data_2d.drop(data_2d.columns[0], axis=1)
mnist = pd.read_csv(data_folder + "mnist.csv")
```

#### Почнемо з датасету data_2d за допомогою методу ліктя визначили, що оптимальна кількість кластерів 2

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data_2d)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.xticks(np.arange(1,11))
plt.title('data_2d')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Тепер проведемо кластерізацію та візуалізуємо результати 

```python
X = data_2d.select_dtypes(include=[np.number])
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
data_2d['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=data_2d['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title('K-means Clustering')
plt.show()
```

#### Тепер кластеризуємо датасет mnist, але спочатку зменшимо кількість ознак до 2

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(mnist)

pca = PCA(n_components=2)
reduced_mnist = pca.fit_transform(df_scaled)
```

```python
df = pd.DataFrame(data = reduced_mnist,
                           columns = ['PC' + str(x) for x in range(1, len(mnist.columns[:2])+1)])
display(df)
```

#### Тепер можна використати метод ліктя та визначити оптимальну кількість кластерів. В даному випадку 3

```python

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.xticks(np.arange(1,11))
plt.title('mnist')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Проведемо кластеризацію та візуалізуємо результат

```python
X = df.select_dtypes(include=[np.number])
```

```python
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(df)

plt.figure(figsize=(8, 6))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=df['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title('K-means Clustering')
plt.show()
```
